```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
reg_scores = pd.read_csv('../results/model_evaluation/reg_scores.csv')
toxic_scores = pd.read_csv('../results/model_evaluation/toxic_scores.csv')
epa_scores = pd.read_csv('../results/model_evaluation/epa_scores.csv')
```

```python
def plot_single_histogram(data, column, title, xlabel, ylabel):
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column], kde=True, bins=15)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
regression_columns = ['RMSE', 'R2', 'MAE', 'MSE']
for col in regression_columns:
    plot_single_histogram(reg_scores, col, 
                          title=f'Distribution of {col} for Regression Models', 
                          xlabel=col, ylabel='Frequency')
```

```python
binary_columns = ['Accuracy', 'Balance Accuracy', 'F1_score', 'MCC', 'AUROC']
for col in binary_columns:
    plot_single_histogram(toxic_scores, col, 
                          title=f'Distribution of {col} for Binary Classifiers', 
                          xlabel=col, ylabel='Frequency')
```

```python
multiclass_columns = ['Accuracy', 'Balance Accuracy', 'F1_score', 'MCC']
for col in multiclass_columns:
    plot_single_histogram(epa_scores, col, 
                          title=f'Distribution of {col} for Multiclass Classifiers', 
                          xlabel=col, ylabel='Frequency')
```
